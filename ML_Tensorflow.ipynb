```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()


# create two constants: a and b
a = tf.constant(4)
b = tf.constant(3)
```

```python
# perform a computation
c = a + b
print(c) # print the shape of c
```

```python
# create a session to execute the graph
session = tf.compat.v1.Session()
```

A execução de ser feita dentro de sessão circunstanciada por "tf.Session" 

image.png

```python
# run the session. It compute the sum
result = session.run(c)
print(result) # print the actual result
```
