```python
# cria um ambiente de CartPole
env = gym.make("CartPole-v1")
# reinicia o ambiente antes de começar
env.reset()

# jogando 10 vezes
for i in range(10):
   #inicializando as variáveis
   done = False
   recompensa_jogo = 0

   while not done:
   #em cada iteração, escolhe uma ação aleatória, executa-a no ambiente com (env.step())  
     action = env.action_space.sample()
     # dá um passo a mais no ambiente (NOVA_OBSERVAÇÃO, RECOMPENSA, FEITO (true or false) E INFORMAÇÃO)
     new_obs, rew, done, info = env.step(action)
    #Recompensa do jogo
     recompensa_jogo += rew
    
    # quando for done, imprime a recompensa acumulativa do jogo e reinicia o ambiente
     if done:
        print('Nova observação %d terminada, recompensa:%d' % (i, recompensa_jogo))
        env.reset()
```

O método step() retorna 4 variárias que fornecem informações sobre a iteração com o ambiente: Observação, Recompensa, Feito e Informações.
**Observação** = um objeto que representa a nova observação (ou estado) do ambiente;
**Recompensa** = número de recompensas obtidas na última ação (tipo float);
**Feito (Done)** = valor booleano (True ou False) usado em tarefas episódicas limitadas em quantidade de iterações. Quando Done = True siginifica que o episódio terminou (tarefa concluída ou agente morreu) e que o ambiente deve ser redefinido;
**Info**= dicionário que fornece informações extras sobre o ambiente, mas que geralmente não são usadas
