```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
tf.reset_default_graph()
```

```python
const1 = tf.constant(3.0, name='constant1')
```

```python
var = tf.get_variable("variable1", shape=[1,2], dtype=tf.float32)
var2 = tf.get_variable("variable2", shape=[1,2], trainable=False, dtype=tf.float32)
```

image.png

```python
op1 = const1 * var
op2 = op1 + var2
op3 = tf.reduce_mean(op2)
```

```python
init= tf.global_variables_initializer() #uma variável precisa ser inicializada
sess = tf.Session()
sess.run(init)
sess.run(op3)
```
