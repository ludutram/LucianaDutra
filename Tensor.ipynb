image.png

```python
a = tf.compat.v1.placeholder(shape=(1,3), dtype=tf.float32)
b = tf.constant([[10,10,10]], dtype=tf.float32)

c = a + b

session = tf.compat.v1.Session()
result = session.run(c, feed_dict={a:[[0.1,0.2,0.3]]})
print(result)
```

```python
import numpy as np

# NB: the first dimension is 'None', meaning that it can be of any lenght
a = tf.compat.v1.placeholder(shape=(None,3), dtype=tf.float32)
b = tf.compat.v1.placeholder(shape=(None,3), dtype=tf.float32)
c = a + b
print(a)
```

```python
sessao = tf.compat.v1.Session()
print(sessao.run(c, feed_dict={a:[[0.1,0.2,0.3]], b:[[10,10,10]]}))
```

```python
v_a = np.array([[1,2,3],[4,5,6]])
v_b = np.array([[6,5,4],[3,2,1]])
print(sessao.run(c, feed_dict={a:v_a, b:v_b}))
```

```python
# variable initialized randomly
var = tf.compat.v1.get_variable("first_variable", shape=[1,3], dtype=tf.float32)
```

```python
# variable initialized with constant values
init_val = np.array([4,5])
var2 = tf.compat.v1.get_variable("second_variable", shape=[1,2], dtype=tf.int32, initializer=tf.constant_initializer(init_val))
```

```python
# create the session
session = tf.compat.v1.Session()
# initialize all the variables
session.run(tf.compat.v1.global_variables_initializer())
```

```python
print(session.run(var))
```

```python
print(session.run(var2))
```
